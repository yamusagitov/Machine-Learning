### а теперь классификация

```python
from sklearn.linear_model import LinearRegression

x = np.linspace(-2, 2, 100)
u = np.random.random(size=100)
y = 1*(0.5*x + u > 0)
plt.scatter(x,y);


model =LinearRegression()

# обучаем нашу модель x-ам и y-ам
model.fit(y.reshape(-1, 1), x.reshape(-1, 1))

xs = np.linspace(-2, 2, 100)
plt.plot(xs, model.coef_[0][0]*xs);

```

```python
x = np.linspace(-5, 5, 300)
u = np.random.random(size=300)
y = 1*(0.5*x + u > 0)
```

```python
from sklearn.linear_model import LinearRegression

model =LinearRegression()
# обучаем нашу модель x-ам и y-ам
model.fit(y.reshape(-1, 1), x.reshape(-1, 1))

sigm = lambda xs: np.exp(xs)/(1 + np.exp(xs))

xs = np.linspace(-5, 5, 100)
plt.plot(xs, sigm (model.coef_[0][0]*xs), color='r');
plt.scatter(x,y);
```

### __Самое частое:__
 - Функция логистического распределения (легко интерпретировать) => логистическая регрессия (логит)
 - Функция нормального распределения => пробит-модель

__Модель:__
- $x$ - регрессоры
- $y \in \{0,1\}$ -целевая переменная

Склонность к тому, чтобы сделать действие $y=1$
$$
y^* = \beta\cdot x
$$

$$
P(y=1) = \frac{e{y^*}{1+e^{y^*}
$$
